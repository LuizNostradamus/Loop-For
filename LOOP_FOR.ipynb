```python
### ATIVIDADE 05 ###
# A ORDEM DO IF E ELIF IMPORTA, PRIMEIRO O MAIOR E DEPOIS O MENOR #
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]

meta = 1200
percentual_bonus = 0.1
bonus_extra = 50
for i in lista_vendas:
  if i >= 2000:
    bonus = percentual_bonus * i + bonus_extra
  elif i >= meta:
    bonus = percentual_bonus * i
  else:
    bonus = 0
  print("Bonus de cada funcionadrio:", bonus)
```

```python
### ATIVIDADE 05 (CORREÇÃO PROFESSOR) ###
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]

meta = 1200
percentual_bonus = 0.1
bonus_extra = 50
for i in lista_vendas:
  if i >= meta:
    bonus = percentual_bonus * i
    if i >= 2000:
      bonus = bonus + bonus_extra
  else:
    bonus = 0
  print("Venda: R$", i, "Bonus: R$", bonus)
```
